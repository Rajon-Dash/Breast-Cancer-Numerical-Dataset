```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X,Y ,test_size=0.2 , random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

# Artificial Neural Network (ANN) Model Implementation

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# #adding the input and first hidden layer
# classifier = Sequential()
# classifier.add(Dense(9,input_dim =29, activation = 'relu'))
# # adding the second hidden layer
# classifier.add(Dense(9,activation = 'relu'))
# # adding the output layer
# classifier.add(Dense(1, activation = 'sigmoid'))




classifier = Sequential()
classifier.add(Dense(256, input_dim=29, activation='relu'))
classifier.add(Dense(128, activation='relu'))
classifier.add(Dense(64, activation='relu'))
classifier.add(Dense(32, activation='relu'))
classifier.add(Dense(16, activation='relu'))
classifier.add(Dense(1, activation='sigmoid'))

```

```python
from tensorflow.keras.utils import plot_model
plot_model(classifier,
           to_file="model.png",
           show_shapes=True,
           show_layer_names=True,
          )



```

```python
classifier.summary()
```

```python
# #compiling  the ANN
# classifier.compile(optimizer='adam',loss = 'binary_crossentropy',metrics=['accuracy'])


from keras.optimizers import Adam
optimizer = Adam(learning_rate=0.000001)
classifier.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#fitting  the ANN
hist = classifier.fit(X_train , Y_train,batch_size=100,epochs=1000)
```

```python
#now testing for test data
Ann_Y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming y_true contains true labels and y_pred contains predicted labels
accuracy = accuracy_score(Y_test, np.round(abs(Ann_Y_pred)))
precision = precision_score(Y_test, np.round(abs(Ann_Y_pred)))
recall = recall_score(Y_test, np.round(abs(Ann_Y_pred)))
f1 = f1_score(Y_test, np.round(abs(Ann_Y_pred)))

print(f'Accuracy: {accuracy*100}')
print(f'Precision: {precision*100}')
print(f'Recall: {recall*100}')
print(f'F1-Score: {f1*100}')
```

# Convolutional Neural Network (CNN) Model Implementation

```python
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense

# Initialize the CNN
cnn = Sequential()

# Step 1: Convolution (1D)
cnn.add(Conv1D(32, 3, activation='relu', input_shape=(X_train.shape[1], 1)))

# Step 2: Pooling (1D)
cnn.add(MaxPooling1D(pool_size=2))

# Step 3: Flattening
cnn.add(Flatten())

# Step 4: Full connection
cnn.add(Dense(units=128, activation='relu'))
cnn.add(Dense(units=1, activation='sigmoid'))

# Compile the model
cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(X_train, Y_train, epochs=1000, batch_size=50, validation_data=(X_test, Y_test))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions
cnn_Y_pred = cnn.predict(X_test)

# Convert to 1D array
cnn_Y_pred = cnn_Y_pred.ravel()

# Calculate metrics
np.round(abs(cnn_Y_pred))
accuracy = accuracy_score(Y_test, np.round(abs(cnn_Y_pred)))
precision = precision_score(Y_test, np.round(abs(cnn_Y_pred)))
recall = recall_score(Y_test, np.round(abs(cnn_Y_pred)))
f1 = f1_score(Y_test, np.round(abs(cnn_Y_pred)))

print(f'Accuracy: {accuracy*100}')
print(f'Precision: {precision*100}')
print(f'Recall: {recall*100}')
print(f'F1-Score: {f1*100}')
```

# Recurrent Neural Network (RNN) Model Implementation

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, RNN
```

```python
# Assuming X_train and X_test are your input data
X_train = X_train.reshape(X_train.shape[0], X_train.shape[2], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[2], 1)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Assuming you have already preprocessed your data and have X_train, y_train, X_test, y_test ready.

# Constructing a more complex LSTM model
model = Sequential()

model.add(LSTM(units=128, activation='relu', input_shape=(29, 1), return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=64, activation='relu', return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(units=1, activation='sigmoid'))

# Applying advanced techniques
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display the model summary
model.summary()

# Train the model
history = model.fit(X_train, Y_train, epochs=1000, batch_size=50, 
                    validation_data=(X_test, Y_test), 
                    callbacks=[early_stopping, reduce_lr])

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, Y_test)
print(f"Test Accuracy: {test_accuracy}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

Rnn_y_pred = model.predict(X_test)
Rnn_y_pred_binary = (Rnn_y_pred > 0.5).astype(int)  # Convert probabilities to binary predictions

# Calculate metrics
np.round(abs(Rnn_y_pred))
accuracy = accuracy_score(Y_test, np.round(abs(Rnn_y_pred)))
precision = precision_score(Y_test, np.round(abs(Rnn_y_pred)))
recall = recall_score(Y_test, np.round(abs(Rnn_y_pred)))
f1 = f1_score(Y_test, np.round(abs(Rnn_y_pred)))

print(f'Accuracy: {accuracy*100}')
print(f'Precision: {precision*100}')
print(f'Recall: {recall*100}')
print(f'F1-Score: {f1*100}')
```
